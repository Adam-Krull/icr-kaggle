All null values in BQ are associated with a negative outcome. I will fill all null values in this column with the median for values in this column associated with the negative outcome.

```python
train.BQ = train.BQ.fillna(train[train.target == 0].BQ.median())
```

```python
train[train.CB.isna()]
```

Not much to learn here, I'll just impute the median of the column.

```python
train.CB = train.CB.fillna(train.CB.median())
```

```python
train[train.CC.isna()]
```

It has one positive and negative result associated with the null values. I'll go ahead and impute the median for the entire column.

```python
train.CC = train.CC.fillna(train.CC.median())
```

```python
train[train.DU.isna()]
```

Only one value, I'll go ahead and impute the median for all values.

```python
train.DU = train.DU.fillna(train.DU.median())
```

```python
train[train.EL.isna()]
```

```python
train[train.EL.isna()].target.value_counts()
```

The proportion of negative to positive outcome roughly matches the proportion of labels in my dataset. I'll move forward imputing the median for the entire column.

```python
train[train.EL.isna()][['EL', 'BQ', 'target']]
```

Interestingly, it looks like where there are NaNs in my EL column, the BQ column has an effect on the target. Where I imputed the median for BQ, it's 0. Where the BQ varies from the imputed values, the target is 1. EL being null doesn't have much impact on my target.

```python
train.EL = train.EL.fillna(train.EL.median())
```

```python
train[train.FC.isna()]
```

Since there is only one missing value for FC, and it's associated with a target of 1, I think I'd better be safe and impute the median for the entire column. There isn't enough information here for me to draw a more meaningful conclusion.

```python
train.FC = train.FC.fillna(train.FC.median())
```

I think at this point, I will continue my trend of imputing the median for all train values for all columns except BQ. With BQ, since all null values were associated with a negative result, I will impute the median for BQ values for negative results only.

```python
train.isna().sum()
```

```python
for col in ['FL', 'FS', 'GL']:
    train[col] = train[col].fillna(train[col].median())
    
train.info()    
```

```python
train = pd.get_dummies(train, columns = ['EJ'])
```

```python
train.columns[1:-3]
```

```python
ttest_df = pd.DataFrame(columns = ['col', 't_value', 'p_value'])

for col in train.columns[1:-3]:
    neg = train[train.target == 0][col]
    pos = train[train.target == 1][col]
    t, p = stats.ttest_ind(neg, pos)
    ttest_df.loc[len(ttest_df.index)] = [col, t, p]
                        
ttest_df.sort_values(by = 'p_value')                        
```
